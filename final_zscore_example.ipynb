# z_score

```python
files=glob('/Users/xuhuimin/Documents/lab/lab data/beijingmobile/busi_exp/*')
```

```python
for f in files:
    name=f.split('/')[-1].split('.')[0]
    data=pd.read_csv(f)
    data['pair']=data['node1'].astype('str')+' '+data['node2'].astype('str')
    diction=Counter(data['pair'].values)
    z_score=pd.read_csv('/Users/xuhuimin/Documents/lab/lab data/beijingmobile/z_score.csv',index_col=0)
    z_score[name]=z_score['pair'].apply(lambda x: diction[x]
                                                    if x in diction
                                                    else np.nan)
    z_score.to_csv('/Users/xuhuimin/Documents/lab/lab data/beijingmobile/z_score.csv')
```

```python
z_score=pd.read_csv('/Users/xuhuimin/Documents/lab/lab data/beijingmobile/z_score.csv',index_col=0)
```

```python
z_score=z_score.fillna(0)
```

```python
mean=z_score.iloc[:,[x for x in range(2,12)]].mean(axis=1)
std=z_score.iloc[:,[x for x in range(2,12)]].std(axis=1)
z_score['z_score']=(z_score['busi_obs']-mean)/std
```
